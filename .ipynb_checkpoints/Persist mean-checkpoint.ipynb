```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10000)
```

```python
df = pd.read_csv('web-traffic-time-series-forecasting/train_2.csv/train_2.csv')\
.melt(id_vars='Page', var_name='date', value_name='Visits')
keys = pd.read_csv('web-traffic-time-series-forecasting/key_2.csv/key_2.csv')
```

```python
keys['Page'] = keys['Page'].str[:-11]
```

```python
df['date'] =  pd.to_datetime(df['date'])
df['date'].max()
```

```python
df = df[df['date']  > (df['date'].max() - pd.Timedelta(days=14))]

df.date.unique()
```

```python
df['Visits'] = df['Visits'].fillna(0)
gr = df.groupby('Page').mean()['Visits'].reset_index(level=0)
subm = pd.merge(keys, gr, how='left', on=["Page"])
```

```python
subm[:5000]
```

```python
subm = subm.drop(columns=['Page'])
```

```python
subm['Visits'] = subm['Visits'].round().astype(int)
subm.to_csv('subm.csv', index=False)
```
